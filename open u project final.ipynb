Background:
LendingClub is a US peer-to-peer lending company, headquartered in San Francisco, California. It was the first peer-to-peer lender to register its offerings as securities with the Securities and Exchange Commission (SEC), and to offer loan trading on a secondary market. LendingClub is the world's largest peer-to-peer lending platform. The company claims that $15.98 billion in loans had been originated through its platform up to December 31, 2015.

LendingClub enables borrowers to create loan listings on its website by supplying details about themselves and the loans that they would like to request. All loans are unsecured personal loans and can be between $1,000 - $40,000. On the basis of the borrower’s credit score, credit history, desired loan amount and the borrower’s debt-to-income ratio, LendingClub determines whether the borrower is credit worthy and assigns to its approved loans a credit grade that determines payable interest rate and fees. The standard loan period is three years; a five-year period is available at a higher interest rate and additional fees. The loans can be repaid at any time without penalty.

Investors can search and browse the loan listings on LendingClub website and select loans that they want to invest in based on the information supplied about the borrower, amount of loan, loan grade, and loan purpose. Investors make money from interest. LendingClub makes money by charging borrowers an origination fee and investors a service fee.

LendingClub also makes traditional direct to consumer loans, including automobile refinance transactions, through WebBank, an FDIC-insured, state-chartered industrial bank that is headquartered in Salt Lake City Utah. The loans are not funded by investors but are assigned to other financial institutions.

Investors make money from interest. Rates vary from 6.03% to 26.06%, depending on the credit grade assigned to the loan request. The grades assigned to these requests range alphabetically from A to G, with A being the highest-grade, lowest-interest loan. Each of these letter grades has five finer-grain sub-grades, numbered 1 to 5, with 1 being the highest sub-grade. LendingClub makes money by charging borrowers an origination fee and investors a service fee. The size of the origination fee depends on the credit grade and ranges to be 1.1%-5.0% of the loan amount. The size of the service fee is 1% on all amounts the borrower pays. The company facilitates interest rates that are better for lenders and borrowers than they would receive from most banks. It has averaged between a six and nine percent return to investors between its founding and 2013. However, because lenders are making personal loans to individuals on the site, their gains are taxable as personal income instead of investment income. Therefore, income from LendingClub loans may be taxed at a higher rate than investments that are taxed at the capital gains rate.

    
### Data preparation for visualization: we will transform variables of type 'object' to numerical when needed
<a id="d"></a>

```python
%%time
df = df.dropna(axis=0, subset=['int_rate'])
df['interest']=df['int_rate']
df['revol']=df['revol_util']

df= df.reset_index(drop=True)
df['int_rate']=df["interest"].apply(lambda x:float(x[:-1])/100 )
df = df.dropna(axis=0, subset=['interest'])

df = df.dropna(axis=0, subset=['revol'])
df= df.reset_index(drop=True)
df['revol_util']=df["revol"].apply(lambda x:float(x[:-1])/100 )


df["emp_length"] = df["emp_length"].replace({'years':'','year':'',' ':'','<':'','\+':'','n/a':'0'}, regex = True)
df["home_ownership"] = df["home_ownership"].map({"OWN":1,"MORTGAGE":2,"RENT":3,"OTHER":4,"ANY":5,"NONE":6})
years = pd.to_datetime(df['issue_d'])
df['year'] = years.dt.year
last_pymnt_date = pd.to_datetime(df['last_pymnt_d'])
df['last_p_year'] = last_pymnt_date.dt.year


df.drop('issue_d', axis=1, inplace=True)
df.drop('interest', axis=1, inplace=True)
df.drop('revol', axis=1, inplace=True)
df.drop('last_pymnt_d', axis=1, inplace=True)

df["term"] = df["term"].replace({'60 months':5,'36 months':3}, regex = True)
df = df.rename(columns= {'emp_length_< 1 year':'emp_length_lt_1 year',
                         'emp_length_n/a':'emp_length_na'})
print("Current shape of dataset :",df.shape)
```

## Defining bad loans
<a id="e"></a>

<p>
<span style="color:blue">
One of the main goal is to try and predict the risky loans. Therefore the definition of bad loans is critical for all stages.
Bad loans - represents loans that with a high probability will not be repaid.
<p>
<span style="color:blue">
(1) We will split the dataframe target/dependent variable/y into good loans and bad loans.
   <p>
<span style="color:blue">   
(2) We define good loans as:'Fully Paid','Current','In Grace Period','Late (16-30 days)'
   <p>
<span style="color:blue">   
(3) We define bad loans data as any loan that is in the "Loan status": "Charged Off","Late (31-120 days)","Default" (close to Basel definition): <p>
    
https://www.bis.org/bcbs/qis/qis3qa_f.htm

<p>
<span style="color:blue">
Loan status

```python
# loan status
df["loan_status"].value_counts()
```

<p>
<span style="color:blue">
(1) Fully Paid - Represent zero risk
    <p>
<span style="color:blue">
(2) Current - Represent loans which were issued and haven’t yet matured
    <p>
<span style="color:blue">
(3) Charged Off -A declaration by a creditor that an amount of debt is unlikely to be collected
    <p>
<span style="color:blue">
(4) Late (31-120 days)  - Represent Loans that were due 31-120 days back, representing high risk
    <p>
<span style="color:blue">
(5) In Grace Period  - Loans that were due yet payment may be made without penalty during this time (usually 15 days)
    <p>
<span style="color:blue">
(6) Late (16-30 days) - Represent Loans that were due 16-30 days back, representing medium risk                                 <p>
<span style="color:blue">    
(7) Does not meet the credit policy. Status: Fully Paid - Represent risky loans that were fully paid
    <p>
<span style="color:blue">
(8) Does not meet the credit policy. Status: Charged Off - represent risky loans that were partially paid
    <p>
<span style="color:blue">
(9) Default - Represent loans that wont be returned                                                    

```python
df['loan_status'] = [1 if (x =='Charged Off' or  x =='Default' or 
                        x =='Late (31-120 days)'or x =="Does not meet the credit policy. Status:Charged Off")
                  else 0 for x in df['loan_status']] 

p=df['loan_status'].sum()/df['loan_status'].count()
print ('The default level is:')
print ("{:.0%}".format(p))
```
